```python

def date2week(row):
    if pd.isnull(row):
        return 0
    date = pd.to_datetime(row,format="%Y%m%d")
    return date.weekday()+1

train_data["f_week"] = train_data["Date_received"].apply(lambda x : date2week(x))

test_data["f_week"] = test_data["Date_received"].apply(lambda x : date2week(x))

# data 仅做数据分析用
data["f_week"] = data["Date_received"].apply(lambda x : date2week(x))
```

#### 2.1.2 优惠券类型

> 优惠券的类型有三种：

- 满减      100:5
- 折扣率    0.95
- 无优惠卷

- 类型，one hot 编码
统一转换为折扣率  = 1 - 5/100

```python
def discount_rate(x):
    x = str(x)
    length = len(x)
    a=1
    b=1
    try:
        if length == 3:
            return 1.0
        elif ":" in x:
            a,b = x.split(":")
            return 1-float(b)/float(a)
    except:
        return x

train_data["f_discount_rate"] = train_data["Discount_rate"].apply(lambda x : discount_rate(x))
test_data["f_discount_rate"] = test_data["Discount_rate"].apply(lambda x : discount_rate(x))

feature_list.append("f_discount_rate")  
```

```python
##  类型
def discount_cls(x):
    x = str(x)
    length = len(x)
    if length == 3:
        return 3
    elif ":" in x:
        return 1
    else:
        return 2

train_data["f_discount_cls"] = train_data["Discount_rate"].copy()
train_data["f_discount_cls"] = train_data["f_discount_cls"].apply(lambda x : discount_cls(x))

test_data["f_discount_cls"] = test_data["Discount_rate"].copy()
test_data["f_discount_cls"] = test_data["f_discount_cls"].apply(lambda x : discount_cls(x))


train_data[["f_cls_jian", "f_cls_discount","f_cls_None"]]=pd.get_dummies(train_data["f_discount_cls"])

test_data[["f_cls_jian", "f_cls_discount","f_cls_None"]]=pd.get_dummies(test_data["f_discount_cls"])


train_data.head()
```

### 3.1 处理标签

"Date" 就是我们要预测的结果，但不是预测日期，而是有没有使用。

 不同情况打上不同标签：
 
- 收到优惠卷，未消费： -1
- 未收到优惠卷，15天内消费： 0
- 收到优惠卷， 消费：  1

```python

def makeLabel(row):
    if pd.isnull(row["Date_received"]) : # and pd.isna(row["Date"]) == False:
        return -1
    elif row["Date"] != "null" :
        res = pd.to_datetime(row["Date"],format="%Y%m%d") - pd.to_datetime(row["Date_received"],format="%Y%m%d")
        if res <= pd.Timedelta(15,"D"):
            return 1
    return 0

data["label"] = data.apply(makeLabel, axis=1)

data["label"].unique()
```

```python
# 性能评价函数
def myauc(test):
    testgroup = test.groupby(['Coupon_id'])
    aucs = []
    for i in testgroup:
        tmpdf = i[1]
        if len(tmpdf['label'].unique()) != 2:
            continue
        fpr, tpr, thresholds = roc_curve(tmpdf['label'], tmpdf['pred'], pos_label=1)
        aucs.append(auc(fpr, tpr))
    return np.average(aucs)
```
